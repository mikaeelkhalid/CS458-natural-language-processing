```
import spacy
```

```
nlp = spacy.load('en_core_web_sm')
```

```
doc = nlp(u'tesla is looking at buy U.S startup for $6 million ')
```

```
#tokenizing with part of speach
for token in doc:
  print(token.text, token.pos)
```

```
#tokenizing with position
for token in doc:
  print(token.text, token.pos_)
```
