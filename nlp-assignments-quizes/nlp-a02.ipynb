```python
path = "/content/drive/My Drive/data/badges.data"
```

```python
corpus = open(path).read()
```

```python
res = re.sub(r'','',corpus)
dlist=res.split('\n')
dlist.remove(dlist[0])
dlist.remove(dlist[-1])

dlist
```

```python
x, y  = [], []
for row in dlist:
    label,name, *rest = row.split(' ')
    x.append(label)
    y.append(name)
```

```python
maxminV = CountVectorizer(min_df = 1, max_df = 1.0, max_features = 8)
simpleV = CountVectorizer()

X = maxminV.fit_transform(y)
Y = simpleV.fit_transform(y)
```

```python
maxminZ = TfidfVectorizer(max_df = 1.0, min_df = 1, max_features = 10)

simpleV = TfidfVectorizer()
A = maxminZ.fit_transform(y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(A, x, shuffle=True, test_size = 0.3, train_size = 0.7)
```

```python
knn = KNeighborsClassifier()
nbb = BernoulliNB()
knn.fit(train_x, train_y)
nbb.fit(train_x, train_y)
```

```python
labels_knn = knn.predict(test_x)
labels_nbc = nbb.predict(test_x)
```

```python
print('KNN: ', accuracy_score(test_y, labels_knn))
print('NBB: ', accuracy_score(test_y, labels_nbc))
```
